```python
def matrisOlustur(x,y):
    return [[random.randint(0, 10) for _ in range(y)] for _ in range(x)]
```

```python
def birimMatrisOlusturucu(x):
    birimMatris = []
    for i in range(x):
        satir = []
        for j in range(x):
            if (i==j):
                satir.append(1)
            else:
                satir.append(0)
        birimMatris.append(satir)
    return birimMatris
                
```

```python
print(birimMatrisOlusturucu(3))
```

```python
def transpozeAl(matris):
    satir_sayisi = len(matris)
    sutun_sayisi = len(matris[0])
    
    transpozeMatris = []
    for i in range(sutun_sayisi):
        satir = []
        for j in range(satir_sayisi):
            satir.append(matris[j][i])
        transpozeMatris.append(satir)
    return transpozeMatris
```

```python
print(transpozeAl(A_Matrisi))
```

```python
def matrisCarpimi(matris1,matris2):
    satir1 = len(matris1)
    sutun1 = len(matris1[0])

    
    satir2 = len(matris2)
    sutun2 = len(matris2[0])



    sonucMatris = []

    if (sutun1 != satir2):

        return "satir sütun sayisi eşit değil carpim yapilamaz"
    
    else:
        for i in range(satir1):
            satir = []
            for k in range(sutun2):
                deger = 0
                for j in range (sutun1):
                    deger += matris1[i][j] * matris2[j][k]
                satir.append(deger)
            sonucMatris.append(satir)
    return sonucMatris
                        
    
```
