```python
import numpy as np
```

```python
def f(x):
    return x**2
```

```python
def turev_al(fonksiyon,x):
    epsilon = 1e-5
    return (fonksiyon(x+epsilon)-fonksiyon(x))/epsilon
    
```

```python
x_noktasi = 3.0
sonuc = turev_al(f, x_noktasi)

print(f"f(x) = x^2 fonksiyonunun x={x_noktasi} noktasındaki türevi:")
print(f"Hesaplanan: {sonuc:.4f}")
print(f"Gerçek (2x): {2 * x_noktasi}")
```

```python
def gradyan_hesapla(fonksiyon,x_vektor):
    epsilon = 1e-5
    grad = np.zeros_like(x_vektor)

    for i in range(len(x_vektor)):
        original_value = x_vektor[i]

        x_vektor[i] = original_value + epsilon
        f_plus = fonksiyon(x_vektor)

        x_vektor[i] = original_value
        f_x = fonksiyon(x_vektor)

        grad[i] = (f_plus - f_x) / epsilon
        
    return grad
        
```

```python
def multivariable_f(v):
    x, y = v[0], v[1]
    return 3 * x**2 + y**2
```

```python
v_noktasi = np.array([2.0, 4.0])
grad_sonuc = gradyan_hesapla(multivariable_f, v_noktasi)

print(f"\nVektörel Fonksiyon Türevi (Gradyan) x={v_noktasi}:")
print(f"Hesaplanan Gradyan: {grad_sonuc}")
print(f"Gerçek Gradyan [6x, 2y]: [{6*2}, {2*4}]")
```

```python
import sympy as sp
```

```python
x,y = sp.symbols('x y')

f = 3*x**2 + y**2

diff_x = sp.diff(f,x)

diff_y = sp.diff(f,y)

print(f"\nFonksiyon: {f}")
print(f"x'e göre türev: {diff_x}") 
print(f"y'ye göre türev: {diff_y}")
```

```python
import torch
```

```python
x = torch.tensor([2.0], requires_grad=True)
y = torch.tensor([4.0], requires_grad=True)
```

```python
f = 3*x**2 + y**2

f.backward()

print("\nPyTorch ile Otomatik Türev:")
print(f"df/dx (x=2 için): {x.grad.item()}") # Çıktı 12.0 olmalı (6*2)
print(f"df/dy (y=4 için): {y.grad.item()}") # Çıktı 8.0 olmalı (2*4)
```

```python
def minBul(fonksiyon,x):
    x_yeni = 0
    learning_rate = 0.1
    print(f"ilk x degeri {x}")
    
    x_yeni = x - learning_rate * turev_al(fonksiyon,x)
    for i in range(100):
        x_yeni = x_yeni - learning_rate * turev_al(fonksiyon,x_yeni)
        print(f"islem sonucu yeni x degeri {x_yeni}")
    return x_yeni
        
```

```python
def fonk(x):
    return x**2 - 4*x + 6
```

```python
print(f"fonksiyon = x**2 - 4*x + 6")

print(f"fonksiyonun minimumu = {minBul(fonk,5)}")
```

```python
def vektorminBul(fonksiyon,baslangic_vektoru):
    mevcut_konum = np.array(baslangic_vektoru, dtype=float)
    learning_rate = 0.01
    
    for i in range(200):
        grad = gradyan_hesapla(fonksiyon, mevcut_konum)
        mevcut_konum = mevcut_konum - (learning_rate * grad)

        if i % 20 == 0:
            print(f"Adım {i}: Konum {mevcut_konum}")

    return mevcut_konum
        
    
```

```python
deneme_noktasi = np.array([10.0, 8.0])

print("\n--- Başlıyor ---")
sonuc = vektorminBul(multivariable_f, deneme_noktasi)
print(f"\nSonuç Konum: {sonuc}")
print("Hedef: [0. 0.]'a çok yakın olmalı.")
```
