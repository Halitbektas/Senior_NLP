```python
import math

def faktoriyel(n):
    if n == 0: return 1
    sonuc = 1
    for i in range(1, n+1):
        sonuc *= i
    return sonuc

def taylor_ile_ex_hesapla(x, terim_sayisi):
    toplam = 0
    
    for n in range(terim_sayisi):
        toplam += x**n/faktoriyel(n) 
        pass 
        
    return toplam

# Test
x_degeri = 3
gercek_deger = math.exp(x_degeri)

print(f"Gerçek e^{x_degeri}: {gercek_deger}")

# Bakalım kaç terimde yaklaşacağız?
for i in [1, 3, 5, 10, 15]:
    tahmin = taylor_ile_ex_hesapla(x_degeri, i)
    hata = abs(gercek_deger - tahmin)
    print(f"{i} Terim ile tahmin: {tahmin:.5f} (Hata: {hata:.5f})")
```
